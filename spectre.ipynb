```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

## 1. Données

On étudie le titrage de l'acide citrique, triacide noté $AH_3$ pour simplifier, par la soude.<br><br>
L'équation de la réaction support de titrage est : 
    $$
    AH_3 + 3OH^{-}\longrightarrow A^{3-} + 3 H_2O
    $$<br>
On note  $c_{titré}$  de concentration molaire apportée en acide citrique en mol/L <br><br>
On note  $c_{titrant}$ de concentration molaire apportée de la soude en mol//L <br><br>
On note $V_{titré}$ le volume de la prise d'essai (en mL)
   <hr>

```python
ctitrant =0.1
ctitre =0.1
Vtitre = 5.0
```

## 2. Calcul des quantités de matières des espèces et tracé des courbes

a) Ecrire la relation à l'équivalence en fonction des données rentrées puis en déduire l'expression du volument équivalent $V_E$<br><br>
**Etude avant l'équivalence**<br><br>
    b) Exprimer le nombre de mol initial de $AH_3$ dans la prise d'essai avant le titrage puis montrer que le nombre de mol de $AH_3$ dans le bécher pour un volume $V_{titrant}$ de solution titrante ajoutée s'exprime : <br>
       $$
    n(AH_3)=c_{titre}.V_{titre}-3.c_{titrant}.V_{titrant}
    $$<br>
    c) Que dire du nombre de mol de $OH^{-}$  avant l'équivalence ? Justifier<br>
    d) Exprimer le nombre de mol de $A^{3-}$ dans le bécher pour un volume $V_{titrant}$ de solution titrante ajoutée.<br>
    e) Exprimer le nombre de mol de $Na^{+}$ dans le bécher pour un volume $V_{titrant}$ de solution titrante ajoutée.<br>
      <hr>
**Etude après l'équivalence**<br><br> 
    </div>

```python
# volume de titrant initial en mL
Vtitrant=0

# calcul du volume équivalent en mL
VE=3*ctitre*Vtitre/ctitrant

#création des listes
nAH3=[]
nOH=[]
nA3=[]
nNa=[]
V=[]

#Bilan de matière avant l'équivalence
while Vtitrant<=VE:
    nbreAH3=ctitre*Vtitre-ctitrant*Vtitrant/3
    nbreOH=0
    nbreA3=ctitrant*Vtitrant
    nbreNa=ctitrant*Vtitrant
    nAH3.append(nbreAH3)
    nOH.append(nbreOH)
    nA3.append(nbreA3)
    nNa.append(nbreNa)
    V.append(Vtitrant)
    Vtitrant=Vtitrant+0.5

#Bilan de matière après l'équivalence
while Vtitrant<=25:
    nbreAH3=0
    nbreOH=ctitrant*Vtitrant-ctitrant*VE
    nbreA3=ctitrant*VE
    nbreNa=ctitrant*Vtitrant
    nAH3.append(nbreAH3)
    nOH.append(nbreOH)
    nA3.append(nbreA3)
    nNa.append(nbreNa)
    V.append(Vtitrant)
    Vtitrant=Vtitrant+0.5

plt.figure (figsize = (10,8))
plt.plot(V,nAH3,marker="+",label="$AH_3$")
plt.plot(V,nOH,marker="x",label="$OH^{-}$")
plt.plot(V,nA3,marker="o",label="$A^{3-}$")
plt.plot(V,nNa,marker="v",label="$Na^{+}$")
plt.xlabel('Volume de solution titrante en mL')
plt.ylabel("Quantité de matière de l'espèce en mmol")
plt.legend()
plt.show()
    
```
